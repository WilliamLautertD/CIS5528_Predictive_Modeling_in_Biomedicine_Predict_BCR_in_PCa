```python
#import
import numpy as np
import pandas as pd
```

```python
#load csv
clinical_df = pd.read_csv("clinical_data.csv")
print(clinical_df.head())
```

```python
print(clinical_df.shape)     # (rows, columns)
print(clinical_df.columns)   # column names
print(clinical_df.info())    # data types
```

```python
#drop unnecessary columns
clinical_df = clinical_df.drop(["primary_gleason", "secondary_gleason", "tertiary_gleason", "time_to_follow-up/BCR", "positive_lymph_nodes", "BCR_PSA"], axis = 1)
print(clinical_df.head())
```

```python
#one-hot encodings
clinical_df = pd.get_dummies(clinical_df, columns = ["pT_stage", "earlier_therapy"], drop_first = True, dtype = int)
print(clinical_df.head())
```

```python
#check which patient id's have missing values
print(clinical_df.loc[
    clinical_df["capsular_penetration"] == "x", 
    "patient_id"
].tolist())

#1122 and 1140 are in our set of 20 patients and have a missing capsular_penetration
```

```python
#remove capsular_penetration since it's missing in 2/20 patients
clinical_df = clinical_df.drop("capsular_penetration", axis = 1)
print(clinical_df.head())
```

```python
#change BCR and lymphovascular invasion from float to int
clinical_df["BCR"] = clinical_df["BCR"].astype("int64")
clinical_df["lymphovascular_invasion"] = clinical_df["lymphovascular_invasion"].astype("int64")
print(clinical_df.head())
```

```python
#define 10 positive and 10 negative cases
bcr_pos_ids = ['1041', '1090', '1100', '1122', '1130', '1137', '1169', '1208', '1219', '1258']
bcr_neg_ids = ['1036', '1060', '1129', '1140', '1179', '1186', '1205', '1264', '1284', '1303']
#change list to integers
bcr_pos_ids = [int(x) for x in bcr_pos_ids]
bcr_neg_ids = [int(x) for x in bcr_neg_ids]
```

```python
#create subset dataframe
pos_df_10 = clinical_df[clinical_df["patient_id"].isin(bcr_pos_ids)]
neg_df_10 = clinical_df[clinical_df["patient_id"].isin(bcr_neg_ids)]
subset_df_20 = pd.concat([pos_df_10, neg_df_10])
```

```python
x = subset_df_20.drop(["BCR", "patient_id"], axis = 1)
y = subset_df_20[["BCR", "patient_id"]]
```

```python
print(y.head())
```

```python
#extract embeddings
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import KFold

import torch
import torch.nn as nn
from torch.utils.data import DataLoader, TensorDataset
```

```python
X_proc = x.copy()
y_proc = y.copy()
X_proc["pre_operative_PSA"] = np.log1p(X_proc["pre_operative_PSA"])
continuous_cols = ["age_at_prostatectomy", "ISUP", "pre_operative_PSA"]
binary_cols = [col for col in X_proc.columns if col not in continuous_cols]
```

```python
class DenoisingAutoencoder(nn.Module):
    def __init__(self, input_dim):
        super().__init__()
        
        self.encoder = nn.Sequential(
            nn.Linear(input_dim, 128),
            nn.ReLU(),
            nn.Linear(128, 64),
            nn.ReLU(),
            nn.Linear(64, 16)
        )
        
        self.decoder = nn.Sequential(
            nn.Linear(16, 64),
            nn.ReLU(),
            nn.Linear(64, 128),
            nn.ReLU(),
            nn.Linear(128, input_dim)
        )

    def forward(self, x):
        z = self.encoder(x)
        out = self.decoder(z)
        return out, z
```

```python
kf = KFold(n_splits=5, shuffle=True, random_state=42)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

embedding_rows = []
```

```python
for fold, (train_idx, val_idx) in enumerate(kf.split(X_proc), start=1):
    print(f"\nStarting fold {fold}")
    
    # Split data
    X_train_df = X_proc.iloc[train_idx].copy()
    X_val_df = X_proc.iloc[val_idx].copy()
    
    y_train_df = y_proc.iloc[train_idx].copy()
    y_val_df = y_proc.iloc[val_idx].copy()
    
    # Fit scaler only on training fold
    scaler = StandardScaler()
    X_train_df[continuous_cols] = scaler.fit_transform(X_train_df[continuous_cols])
    X_val_df[continuous_cols] = scaler.transform(X_val_df[continuous_cols])
    
    # Convert to numpy
    X_train_array = X_train_df.values.astype(np.float32)
    X_val_array = X_val_df.values.astype(np.float32)
    
    # Datasets / loaders
    train_dataset = TensorDataset(torch.tensor(X_train_array))
    val_dataset = TensorDataset(torch.tensor(X_val_array))
    
    train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True)
    val_loader = DataLoader(val_dataset, batch_size=16, shuffle=False)
    
    # Model
    input_dim = X_train_array.shape[1]
    model = DenoisingAutoencoder(input_dim).to(device)
    
    # Column indices
    cont_idx = [X_train_df.columns.get_loc(col) for col in continuous_cols]
    bin_idx = [X_train_df.columns.get_loc(col) for col in binary_cols]
    
    # Losses and optimizer
    mse_loss = nn.MSELoss()
    bce_loss = nn.BCEWithLogitsLoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)
    
    def compute_loss(recon, target):
        recon_cont = recon[:, cont_idx]
        target_cont = target[:, cont_idx]
        
        recon_bin = recon[:, bin_idx]
        target_bin = target[:, bin_idx]
        
        loss_cont = mse_loss(recon_cont, target_cont)
        loss_bin = bce_loss(recon_bin, target_bin)
        
        total_loss = loss_cont + loss_bin
        return total_loss, loss_cont, loss_bin
    
    # Early stopping settings
    best_val_loss = float("inf")
    best_model_state = None
    patience = 20
    patience_counter = 0
    num_epochs = 300
    noise_std = 0.1
    
    # Training loop
    for epoch in range(num_epochs):
        model.train()
        train_loss = 0.0
        
        for (batch,) in train_loader:
            batch = batch.to(device)
            
            noisy_batch = batch + noise_std * torch.randn_like(batch)
            
            optimizer.zero_grad()
            recon, _ = model(noisy_batch)
            loss, _, _ = compute_loss(recon, batch)
            loss.backward()
            optimizer.step()
            
            train_loss += loss.item()
        
        train_loss /= len(train_loader)
        
        model.eval()
        val_loss = 0.0
        
        with torch.no_grad():
            for (batch,) in val_loader:
                batch = batch.to(device)
                recon, _ = model(batch)
                loss, _, _ = compute_loss(recon, batch)
                val_loss += loss.item()
        
        val_loss /= len(val_loader)
        
        if val_loss < best_val_loss:
            best_val_loss = val_loss
            best_model_state = model.state_dict()
            patience_counter = 0
        else:
            patience_counter += 1
        
        if epoch % 20 == 0:
            print(f"Fold {fold}, Epoch {epoch}: train={train_loss:.4f}, val={val_loss:.4f}")
        
        if patience_counter >= patience:
            print(f"Early stopping on fold {fold} at epoch {epoch}")
            break
    
    # Load best model for this fold
    model.load_state_dict(best_model_state)
    model.eval()
    
    # Extract embeddings for validation patients only
    X_val_tensor = torch.tensor(X_val_array).to(device)
    
    with torch.no_grad():
        _, val_embeddings = model(X_val_tensor)
    
    val_embeddings = val_embeddings.cpu().numpy()
    
    # Save rows
    for i, row_idx in enumerate(val_idx):
        row_dict = {
            "patient_id": y_proc.iloc[row_idx]["patient_id"],
            "BCR": y_proc.iloc[row_idx]["BCR"],
            "fold": fold
        }
        
        for j in range(16):
            row_dict[f"embed_{j+1}"] = val_embeddings[i, j]
        
        embedding_rows.append(row_dict)
```

```python
embeddings_df = pd.DataFrame(embedding_rows)

# optional: sort by patient_id
embeddings_df = embeddings_df.sort_values("patient_id").reset_index(drop=True)

print(embeddings_df.head())
print(embeddings_df.shape)
```
